```python
import pandas as pd
import os
import shutil


images_source = 'data/low/2022'
output = 'labeled_sets/labeled_low/'
splitted_output = 'labeled_splits/split_low'

df = pd.read_csv('is_cloudy.csv', dtype={'year': str, 'time': str})

print(df.head()) 
```

```python
blacklist_date_low = ["20220101","20220102","20220103","20220104","20220105","20220106","20220107","20220108","20220109",
    "20220110","20220111","20220112","20220113","20220114","20220115","20220116","20220117","20220118","20220119","20220120","20220121"
    ,"20220122","20220123","20220124","20220125","20220126",
    "20220217","20220218","20220224","20220225","20220226","20220227","20220305","20220306","20220307","20220308",
    "20220311","20220321","20220329","20220330","20220620","20220621","20220622","20220623","20220624","20220625",
    "20220626","20220627","20220628","20220629","20220630","20220631","20220701","20221007","20221209","20221223",
    "20221224","20221225","20221231"
]
```

```python
# Change blacklist for different sites
blacklist_date = blacklist_date_low
# blacklist_date = blacklist_date_blo


date_list = df.year.unique()
```

```python
for filename in os.listdir(images_source):
    names = filename.split('-')
    date_string = names[3]
    time_string = (names[4].split('.'))[0]
    
    if date_string not in date_list:
        continue
    if  date_string in blacklist_date:
        continue
        
    is_cloudy = df.loc[(df['year'] == date_string) & (df['time'] == time_string), 'isCloudy'].values
    if (is_cloudy.size == 0):
        continue
    f = os.path.join(images_source, filename)
    if is_cloudy.any():
        print("cloudy")
        shutil.copyfile(f, output + 'cloudy/' + filename)
    else:
        print("clear")
        shutil.copyfile(f, output + 'clear/' + filename)
    
    
```
